```python
import math
import matplotlib.pyplot as plt
import numpy as np
import random

f=lambda X: X[0]**2+X[1]**2
df= lambda X: [2*X[0],2*X[1]]

#Preparando datos para mapa de nivel
resolucion=30
rango=2.5
X=np.linspace(-rango,rango,resolucion)
Y=np.linspace(-rango,rango,resolucion)
Z=np.zeros((resolucion,resolucion))

for ix, x in enumerate (X):
  for iy, y in enumerate (Y):
    Z[ix,iy]=f([x,y])


#Grafica de mapa de nivel Z
plt.contourf(X,Y,Z,resolucion)
plt.colorbar()

#Punto aleatorio inicial
P=[random.uniform(-2,2),random.uniform(-2,2)]
plt.plot(P[0],P[1],"o",c="white")

#Tasa de aprendizaje
TA=0.1

#Iteraciones
for _ in range(500):
  grad=df(P)

  P[0],P[1]=P[0]-TA*grad[0], P[1]-TA*grad[1]
  plt.plot(P[0],P[1],"o",c="red")

plt.plot(P[0],P[1],"o",c="green")
plt.show()
print("Solucion:", P, f(P))
```

```python
f= lambda X: np.sin(1/2*X[0]**2-1/4*X[1]**2+3)*np.cos(2*X[0]+1-np.e**X[1])

#Preparando datos para mapa de nivel
resolucion=30
rango=2.5
X=np.linspace(-rango,rango,resolucion)
Y=np.linspace(-rango,rango,resolucion)
Z=np.zeros((resolucion,resolucion))

for ix, x in enumerate (X):
  for iy, y in enumerate (Y):
    Z[ix,iy]=f([x,y])


#Grafica de mapa de nivel Z
plt.contourf(X,Y,Z,resolucion)
plt.colorbar()


def df(Punto):
  h=0.01
  T=np.copy(Punto)
  grad=np.zeros(2)
  for it, th in enumerate (Punto):
    T[it]=T[it]+h
    grad[it]=(f(T)-f(Punto))/h
  return grad


P=[random.uniform(-2,2),random.uniform(-2,2)]

#Tasa de aprendizaje
TA=0.1

#Iteraciones
for _ in range(500):
  grad=df(P)

  P[0],P[1]=P[0]-TA*grad[0], P[1]-TA*grad[1]
  plt.plot(P[0],P[1],"o",c="red")

plt.plot(P[0],P[1],"o",c="green")
plt.show()
print("Solucion:", P, f(P))
```
